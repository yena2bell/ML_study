```python
import numpy as np

x = np.array([[1,2,3],[4,5,6]])
print("x:\n{}".format(x))
```

```python
from scipy import sparse

#대각선 원소가 1이고 나머지는 0인 2차원 numpy 행렬
eye = np.eye(4)
print("Numpy 배열:\n{}".format(eye))

#identity도 identity matrix를 만든다. 
#eye는 k에 값을 부여해서 변형이 가능.

eye_kis1 = np.eye(4,k=1)
print("when k is 1:\n{}".format(eye_kis1))
```

```python
#numpy 배열을 CSR 포맷의 SciPy 희소 행렬로 변환
#0이 아닌 원소만 저장한다.
sparse_matrix = sparse.csr_matrix(eye)
print("ScyPy의 CSR 행렬:\n{}".format(sparse_matrix))
```

```python
#희소 행렬을 만드는 다른 방법.
data= np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO 표현:\n{}".format(eye_coo))
```

```python
print(type(eye_coo) == type(sparse_matrix))
#흠 같은 값이지만 세세한 type까지는 다른 객체로군.
```
